```python
import os
import cv2
import numpy as np

path_to_movie_1 = os.path.join('..','src','data','plume_movies','intermittent_smoke.avi')
path_to_movie_2 = os.path.join('..','src','data','plume_movies','new_smoke_1a.avi')
```

```python
video_capture_1 = cv2.VideoCapture(path_to_movie_1)
video_capture_2 = cv2.VideoCapture(path_to_movie_2)
fps_1 = video_capture_1.get(cv2.CAP_PROP_FPS)
fps_2 = video_capture_2.get(cv2.CAP_PROP_FPS)
```

```python
video_capture_2.set(cv2.CAP_PROP_POS_FRAMES, 999)
_, frame_999 = video_capture_2.read()
frame_999_r = frame_999[:, :, 2].T
video_capture_2.set(cv2.CAP_PROP_POS_FRAMES, 1000)
_, frame_1000 = video_capture_2.read()
frame_1000_r = frame_1000[:, :, 2].T
video_capture_2.set(cv2.CAP_PROP_POS_FRAMES, 1001)
_, frame_1001 = video_capture_2.read()
frame_1001_r = frame_1001[:, :, 2].T
video_capture_2.set(cv2.CAP_PROP_POS_FRAMES, 1002)
_, frame_1002 = video_capture_2.read()
frame_1002_r = frame_1002[:, :, 2].T
video_capture_2.set(cv2.CAP_PROP_POS_FRAMES, 1003)
_, frame_1003 = video_capture_2.read()
frame_1003_r = frame_1003[:, :, 2].T
video_capture_2.set(cv2.CAP_PROP_POS_FRAMES, 1004)
_, frame_1004 = video_capture_2.read()
frame_1004_r = frame_1004[:, :, 2].T
video_capture_2.set(cv2.CAP_PROP_POS_FRAMES, 1005)
_, frame_1005 = video_capture_2.read()
frame_1005_r = frame_1005[:, :, 2].T
```
