```python
import numpy as np 
import matplotlib.pyplot as plt 

reward_hist_array = np.zeros((10,2000))

path = '../trained_models/new_movie_with_int_061923/normal_reward/'


def smooth(data, num_pts, mode):

    box = np.ones(num_pts)/num_pts
    y_smooth = np.convolve(data, box, mode=mode)

    return y_smooth

for i in range(0,10):
    
    s = np.load(path+str(i)+"_success_history.npy")
    print(np.mean(s[0:10]))
    print(np.mean(s[0:400]))
    s = smooth(s, num_pts = 100, mode = 'valid')
    print(i)
    print(len(s))
    print(np.sum(s))
    print(np.mean(s[6000:]))
    print(np.max(s))
    #print(np.sum(s[0:150]))
    
    x = np.arange(len(s))  
    plt.plot(x,s)
    plt.ylim([0,0.5])
    #plt.axvline(lines, c='r')

    plt.show()    
    

```
